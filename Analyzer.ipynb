```python
import pandas as pd

# Load your dataset
file_path = 'C:/Users/Kavya/OneDrive/Desktop/Bsd.csv'
data = pd.read_csv(file_path)

def generate_boolean_query(df, diagnosis=None, radius_range=None, area_range=None):
    query = []

    if diagnosis:
        query.append(f"diagnosis == '{diagnosis}'")

    if radius_range:
        min_radius, max_radius = radius_range
        query.append(f"radius_mean >= {min_radius} and radius_mean <= {max_radius}")

    if area_range:
        min_area, max_area = area_range
        query.append(f"area_mean >= {min_area} and area_mean <= {max_area}")

    boolean_query = " and ".join(query)

    # Filter dataset based on the generated boolean query
    filtered_data = df.query(boolean_query) if boolean_query else df

    return boolean_query, filtered_data

# Define parameters for the query
diagnosis = 'M'  # Malignant
radius_range = (15, 25)  # Example radius range
area_range = (500, 1500)  # Example area range

# Generate and execute the Boolean query
query, filtered_data = generate_boolean_query(data, diagnosis, radius_range, area_range)

# Output the query and filtered data
query, filtered_data.head()
```

```python
import pdfplumber
import re
from collections import Counter

def extract_text_from_pdf(pdf_path):
    """Extracts and returns text content from a PDF file."""
    text = ""
    with pdfplumber.open(pdf_path) as pdf:
        for page in pdf.pages:
            text += page.extract_text() + "\n"
    return text

def keyword_extraction(text, keywords):
    """Extracts keywords from text and returns a frequency count."""
    # Normalize text to lowercase for case-insensitive matching
    text_lower = text.lower()
    keyword_counts = Counter()

    for keyword in keywords:
        # Search for the keyword in the text
        occurrences = len(re.findall(r'\b' + re.escape(keyword.lower()) + r'\b', text_lower))
        if occurrences > 0:
            keyword_counts[keyword] = occurrences

    return keyword_counts

# Define the path to your PDF file
pdf_path = "E:/healthcare-chatbot/Sample Training Report Template.pdf"

# List of keywords to search for in the PDF text
keywords = [
    'cancer', 'tumor', 'diagnosis', 'malignant', 'benign', 'treatment', 'therapy',
    'chemotherapy', 'radiation', 'biopsy', 'screening', 'symptoms', 'prognosis',
    'risk', 'genetics', 'immunotherapy', 'metastasis', 'oncology'
]

# Extract text from the PDF
pdf_text = extract_text_from_pdf(pdf_path)

# Debugging Step: Print a sample of the extracted text to verify
print("Sample Text Contains :\n", pdf_text[:500])  # Print the first 500 characters

# Perform keyword extraction
keyword_counts = keyword_extraction(pdf_text, keywords)

# Check if any keywords were found and output results
if keyword_counts:
    print("Detected Keywords and Frequencies:")
    for keyword, count in keyword_counts.items():
        print(f"{keyword}: {count}")
else:
    print("No disease-positive data detected.")
    print("Hence he/she is free from infection")
```
